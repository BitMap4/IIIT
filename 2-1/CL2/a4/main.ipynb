```python
from Classifier   import Classifier
from read_reviews import read_reviews, count_reviews
from itertools    import islice
```

```python
p       = int(0.8 * count_reviews("movie_reviews/pos"))
n       = int(0.8 * count_reviews("movie_reviews/neg"))
p_iter  = read_reviews("movie_reviews/pos")
n_iter  = read_reviews("movie_reviews/neg")
p_train = islice(p_iter, p)
n_train = islice(n_iter, n)
C1      = Classifier({
    "pos": p_train,
    "neg": n_train
}, {
    "pos": p,
    "neg": n
})
```

```python
C1.feature_counts()
```

```python
for review in p_iter:
    print(C1.classify(review))
```

```python
for review in n_iter:
    print(C1.classify(review))
```
